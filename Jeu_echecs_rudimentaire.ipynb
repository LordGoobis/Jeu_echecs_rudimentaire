```python
def chercherPosPiece(m, piece):
    assert len(piece) == 3
    for i in range(len(m)):
        for j in range(len(m[0])):
            if piece == m[i][j]:
                return(i,j)  # on renvoie les coordonnées. i = ligne (en partant du haut), j  = colonne (en partant de la gauche)
    return False #Si on ne trouve pas la position de la pièce, on retourne False
```
